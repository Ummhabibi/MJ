#### Select your strategy to handle missing values, and tell us why you had made that choice.

Two columns have missing values; 'reading_remarque' and 'counter_number'. 

The 'reading_remarque' column has a lot of missing values(4531), ergo filling in the missing values may introduce significant bias. As such, dropping the entire column is the best option.

The 'counter_number' has fewer missing values(48); and since the data type is float, I'd opt to fill in the missing values with the mean value of the column.








```python
#Dropping 'reading_remarque' column

df.drop(columns=['reading_remarque'], inplace=True)
```

```python
#Verify column drop
df.info()
```

```python
#Fill missing values in the column with the mean value of the column

df['counter_number'].fillna(df['counter_number'].mean(), inplace=True)
```

```python
#Verify change
print(df.isnull().sum())
```
